## finding the optimal time for advertisement 

```python

df1=data_set[['beginTime','endTime','duration','totalViewerDuration','avgViewerDuration']]
df1
```

```python
df1['beginTime']=pd.to_datetime(df1['beginTime'])
df1['beginhours']=df1['beginTime'].dt.hour
df1['endTime']=pd.to_datetime(df1['endTime'])
df1['endhours']=df1['endTime'].dt.hour
df1
```

```python
df1.drop(['beginTime','endTime','end'],axis=1)
```

```python
plt.bar(df1['beginhours'],df1['avgViewerDuration'])
plt.xlabel('Begin_hours')
plt.ylabel('avg viewer duration')
plt.title('viewer duration VS hours of the day')
plt.xticks(range(0,24))
plt.tight_layout()
plt.show()
```

### as per the observation from the data,,,the optimal time for the placement of the advertisements is "12am,3pm,7pm"
